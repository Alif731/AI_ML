Splitting dataset

```python
# Split train into X_train (Features) and y_train (Target) :
x=data.drop(['ID','default payment next month' ],axis=1)
y=data['default payment next month']
```

```python
#  scale X_train and y_train with robust scaler
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_sc = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_sc,y,test_size=0.25,random_state=42)
```

```python
# training with models such as kNN, SVM, Gradient Boosting,  Ada boosting and XG Boost

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score

# Train and evaluate KNN
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
f1_score_knn = f1_score(y_test, y_pred_knn)

# Train and evaluate SVM
svm = SVC()
svm.fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
f1_score_svm = f1_score(y_test, y_pred_svm)

# Train and evaluate Gradient Boosting
gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
y_pred_gb = gb.predict(x_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
f1_score_gb = f1_score(y_test, y_pred_gb)

# Train and evaluate AdaBoost
ada = AdaBoostClassifier()
ada.fit(x_train, y_train)
y_pred_ada = ada.predict(x_test)
accuracy_ada = accuracy_score(y_test, y_pred_ada)
f1_score_ada = f1_score(y_test, y_pred_ada)

# Train and evaluate XG Boost
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
f1_score_xgb = f1_score(y_test, y_pred_xgb)

# Print accuracy and f1 score for each model
print("KNN Accuracy:", accuracy_knn)
print("KNN F1 Score:", f1_score_knn)
print("SVM Accuracy:", accuracy_svm)
print("SVM F1 Score:", f1_score_svm)
print("Gradient Boosting Accuracy:", accuracy_gb)
print("Gradient Boosting F1 Score:", f1_score_gb)
print("AdaBoost Accuracy:", accuracy_ada)
print("AdaBoost F1 Score:", f1_score_ada)
print("XG Boost Accuracy:", accuracy_xgb)
print("XG Boost F1 Score:", f1_score_xgb)
```

```python
#cross validating each model with stratified

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
# Perform cross-validation for each model
cv_scores_knn = cross_val_score(knn, x_sc, y, cv=skf, scoring='f1')
cv_scores_svm = cross_val_score(svm, x_sc, y, cv=skf, scoring='f1')
cv_scores_gb = cross_val_score(gb, x_sc, y, cv=skf, scoring='f1')
cv_scores_ada = cross_val_score(ada, x_sc, y, cv=skf, scoring='f1')
cv_scores_xgb = cross_val_score(xgb, x_sc, y, cv=skf, scoring='f1')

# Print the average cross-validation scores
print("Average KNN CV Score:", cv_scores_knn.mean())
print("Average SVM CV Score:", cv_scores_svm.mean())
print("Average Gradient Boosting CV Score:", cv_scores_gb.mean())
print("Average AdaBoost CV Score:", cv_scores_ada.mean())
print("Average XG Boost CV Score:", cv_scores_xgb.mean())
```

```python
# Grid search through Gradient Boosting Model

from sklearn.model_selection import GridSearchCV

# Define the grid of hyperparameters
param_grid = {
    'learning_rate': [0.1, 0.2],
    'n_estimators': [150, 200],
    'max_depth': [4, 5],
    'subsample': [0.7, 0.8],
    'random_state': [0, 42]
}

# Create the GridSearchCV object
grid_gb = GridSearchCV(gb, param_grid, cv=5, n_jobs=-1)

# Fit the model
grid_gb.fit(x_train, y_train)

# Print the best parameters
print("Best parameters:", grid_gb.best_params_)
print("Best Score:", grid_gb.best_score_)

# Predict on the test set
y_pred_gb_grid = grid_gb.predict(x_test)

# Evaluate the model
accuracy_gb_grid = accuracy_score(y_test, y_pred_gb_grid)
f1_score_gb_grid = f1_score(y_test, y_pred_gb_grid)

# Print the accuracy and f1 score
print("Gradient Boosting with Grid Search Accuracy:", accuracy_gb_grid)
print("Gradient Boosting with Grid Search F1 Score:", f1_score_gb_grid)
```

```python
# randomized search through Gradient Boosting Model

from sklearn.model_selection import RandomizedSearchCV

# Define the hyperparameter space
params = {
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [3, 4, 5, 6, 8, 10, 12, 15],
    'subsample': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
    'random_state': [0, 42, 101]
}

# RandomizedSearchCV object
rs_gb = RandomizedSearchCV(gb, params, cv=5, n_iter=5, random_state=0, n_jobs=-1)

# Fitting the model
rs_gb.fit(x_train, y_train)

# Print the best parameters
print("Best parameters:", rs_gb.best_params_)
print("Best Score:", rs_gb.best_score_)

# Predict on the test set
y_pred_gb_rs = rs_gb.predict(x_test)

# Evaluate the model
accuracy_gb_rs = accuracy_score(y_test, y_pred_gb_rs)
f1_score_gb_rs = f1_score(y_test, y_pred_gb_rs)

# Print the accuracy and f1 score
print("Gradient Boosting with Randomized Search Accuracy:", accuracy_gb_rs)
print("Gradient Boosting with Randomized Search F1 Score:", f1_score_gb_rs)
```

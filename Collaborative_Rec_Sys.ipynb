```python
from surprise import Dataset, Reader, KNNWithMeans, SVD
from surprise.accuracy import rmse
```

```python
reader = Reader(rating_scale=(-10, 10))
```

```python
data = Dataset.load_from_df(df, reader)
```

```python
from surprise.model_selection import cross_validate
```

```python
# KNNWithMeans algorithm
sim_options = {'name': 'pearson_baseline', 'user_based': False}
algo = KNNWithMeans(sim_options=sim_options)
cross_validate(algo,data,measures=['RMSE'],cv=5,verbose=True)
```

```python
# SVD algorithm
algo1= SVD()
cross_validate(algo1,data,measures=['RMSE'],cv=5,verbose=True)
```

```python
# Predicting the value of rating based on best algorithm
user = 11765
jokeid = 2432
pred = algo.predict(user,jokeid)
pred.est
```

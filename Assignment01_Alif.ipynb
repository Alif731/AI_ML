Importing python  libraries like numpy, pandas and matplotlib for working with the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Reading the dataset into the python environment.

```python
data = pd.read_csv('diabetes.csv')
```

2. Displaying the columns in the dataset.

```python
for column in data.columns:
  print(column)
```

3. Calculating the mean of each column of the dataset.

```python
data.mean(axis=0)
```

4. Checking for the null values present in the dataset.

```python
data.isna().sum()
```

5.Visualizing the dataset using matplotlib.

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['Glucose'],data['Insulin'])
plt.title('Glucose Level vs Insulin Level')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
#Plot shows comparison between Glucose level and Insulin
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data['BMI'],color='blue',rwidth=0.95)
plt.title('BMI')
plt.show()
#Plot shows normal range of BMI
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['BMI'],data['Insulin'])
plt.title('BMI vs Insulin')
plt.xlabel('BMI')
plt.ylabel('Insulin')
plt.show()
#Plot shows relation BMI and Insulin
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['BloodPressure'])
plt.xlabel("Blood Pressure")
plt.ylabel("Value")
plt.title("Blood Pressure")
plt.show()
#Boxplot shows Blood pressure's normal range and people who are at risk(outliers)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=data['Pregnancies'])
plt.show()
#Plot shows the number of Pregnancies
```

```python
pregnancies = data["Pregnancies"]
insulin = data["Insulin"]
plt.figure(figsize=(6, 4))
plt.bar(pregnancies, insulin)
plt.xlabel("Pregnancies")
plt.ylabel("Insulin")
plt.title("Insulin by Pregnancies")
plt.xticks(rotation=90)
plt.show()
# Data shows that as number of pregnancies increases, insulin decreases
```

```python
sns.pairplot(data[["Insulin","Glucose","DiabetesPedigreeFunction"]])
plt.show()
#Relation between Insulin, Glucose and Diabetes Pedigree Function
```

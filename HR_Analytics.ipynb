```python
# Split train into X_train (Features) and y_train (Target) :
x=train.drop(['employee_id','is_promoted'],axis=1)
y=train['is_promoted']
```

```python
# prompt: scale X_train and y_train with robust scaler

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_sc = scaler.fit_transform(x)
```

```python
#  robust scale test without 'employee_id' cloumn

x_test = test.drop(['employee_id'], axis=1)
x_test_sc = scaler.transform(x_test)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_sc,y,test_size=0.25,random_state=42)
```

```python
# prompt: train the model with RandomForest , AdaBoost , Gradient Boosting and XGBoost

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
# Random Forest
rfc = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
rfc.fit(x_train, y_train)

rfc_pred= rfc.predict(x_test)
# AdaBoost
abc = AdaBoostClassifier(n_estimators=100, random_state=0)
abc.fit(x_train, y_train)

abc_pred= abc.predict(x_test)
# Gradient Boosting
gbc = GradientBoostingClassifier(n_estimators=100, random_state=0)
gbc.fit(x_train, y_train)

gbc_pred= gbc.predict(x_test)
# XGBoost
xgb = XGBClassifier(max_depth=5,learning_rate=0.3,)
xgb.fit(x_train, y_train)

xgb_pred= xgb.predict(x_test)
```

```python
#  print accuracy and f1 score above models

from sklearn.metrics import accuracy_score, f1_score
print("Random Forest Accuracy:", accuracy_score(rfc_pred, y_test))
print("Random Forest F1 Score:", f1_score(rfc_pred, y_test))
print("AdaBoost Accuracy:", accuracy_score(abc_pred, y_test))
print("AdaBoost F1 Score:", f1_score(abc_pred, y_test))
print("Gradient Boosting Accuracy:", accuracy_score(gbc_pred, y_test))
print("Gradient Boosting F1 Score:", f1_score(gbc_pred, y_test))
print("XGBoost Accuracy:", accuracy_score(xgb_pred, y_test))
print("XGBoost F1 Score:", f1_score(xgb_pred, y_test))
```

```python
# cross validating  RandomForest , AdaBoost , Gradient Boosting and XGBoost

from sklearn.model_selection import cross_val_score

# Random Forest
rfc_cv_scores = cross_val_score(rfc, x_sc, y, cv=5, scoring='f1')
print("Random Forest Cross Validation Mean f1 score:", rfc_cv_scores.mean())

# AdaBoost
abc_cv_scores = cross_val_score(abc, x_sc, y, cv=5, scoring='f1')
print("AdaBoost Cross Validation Mean f1 score:", abc_cv_scores.mean())

# Gradient Boosting
gbc_cv_scores = cross_val_score(gbc, x_sc, y, cv=5, scoring='f1')
print("Gradient Boosting Cross Validation Mean f1 sccore:", gbc_cv_scores.mean())

# XGBoost
xgb_cv_scores = cross_val_score(xgb, x_sc, y, cv=5, scoring='f1')
print("XGBoost Cross Validation Mean f1 score:", xgb_cv_scores.mean())
```

```python
from sklearn.model_selection import  GridSearchCV
# Hyperparameter tuning with Grid Search
param_grid = {
     'max_depth':[5,8],
       'verbosity':[2],
          'learning_rate': [0.1,0.3,]
}

grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, n_jobs=-1, verbose=0)
grid_search.fit(x_train, y_train)

print("Best Parameters found by Grid Search:")
print(grid_search.best_params_)

# Evaluate on the test set with best parameters from Grid Search
best_xgb = grid_search.best_estimator_
y_pred1 = best_xgb.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred1)
print("Test Accuracy with Best Parameters from Grid Search:", test_accuracy)
```
